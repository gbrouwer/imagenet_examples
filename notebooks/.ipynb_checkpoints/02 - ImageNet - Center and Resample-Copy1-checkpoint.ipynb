### Get Bounding Box

### Crop

```python
def crop(im,xmin,ymin,xmax,ymax):

    #Crop
    width = ymax-ymin
    height = xmax-xmin
    centery = int((ymax-ymin)/2) + int(ymin)
    centerx = int((xmax-xmin)/2) + int(xmin)
    if (width > height):
        extent = int((ymax-ymin)/2)
    else:
        extent = int((xmax-xmin)/2)
    ymin = centery-extent
    ymax = centery+extent
    xmin = centerx-extent
    xmax = centerx+extent
    if (ymin < 0):
        ymin = 0
    if (xmin < 0):
        xmin = 0
    if (ymax > imorig.shape[0]):
        ymax = imorig.shape[0]
    if (xmax > imorig.shape[1]):
        xmax = imorig.shape[1]
    cropped = imorig.copy()
    if (len(cropped.shape) > 2):
        cropped = cropped[ymin:ymax-1,xmin:xmax-1,:]
    else:
        cropped = cropped[ymin:ymax-1,xmin:xmax-1]

    #Compute Squarenes
    v1 = im.shape[0]
    v2 = im.shape[1]
    ratio = float(v1)/float(v2)
    ratio_cutoff = 1.2
    if (ratio < ratio_cutoff and ratio > (1/ratio_cutoff)):
        passes = True
    else:
        passes = False    
    
    #Return
    return cropped,ratio,passes
```

### Square Centered Image Creation

```python
columns = imagelist.columns
targetsize = 64
new = []
for i in tqdm(range(len(imagelist))):
    
    #Load Image
    dp = imagelist.iloc[i]
    imorig = imageio.imread(dp.imgfiles)
    
    #Get Image + Bounding Box
    imbb,xmin,ymin,xmax,ymax = boundingbox(imorig,dp)
    
    #Crop
    cropped,squareness,passes = crop(imorig,xmin,ymin,xmax,ymax)

    #If Square, resize to targetsize
    if (squareness == 1.0 and cropped.shape[0] > 64 and cropped.shape[1] > 64):

        #Resize
        resized = transform.resize(cropped, (targetsize,targetsize), \
                        mode='symmetric', preserve_range=True).astype('uint8')
        
        #Store
        storename = dp.imgfiles.replace('/data/','/data/resampled' + str(targetsize) + 'x' + str(targetsize) + '/')
        storepath = storename.split('/')
        storepath = '/'.join(storepath[:-1])
        os.system('mkdir -p ' + storepath)
        new.append((dp.code,dp.shortname,dp.name,dp.description,storepath,storename))
        imageio.imwrite(storename,resized)

new = pd.DataFrame(new)
new.columns = ['code','shortname','name','dp.description','path','imgfile']
new.to_csv('../data/csv/resampled64x64.csv',index=False)
```
